```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

#as colunas "texto" e "rotulo" (0 para não ser do autor, 1 para ser do autor)

# Divida o DataFrame em dados de treinamento e teste
dados_treinamento, dados_teste = train_test_split(df_final1, test_size=0.2, random_state=42)

# Pré-processamento dos textos
max_palavras = 10000  # Número máximo de palavras a serem consideradas no vocabulário
max_comprimento_texto = 200  # Tamanho máximo de cada texto
tokenizer = Tokenizer(num_words=max_palavras, oov_token='<OOV>')
tokenizer.fit_on_texts(dados_treinamento["text"])
sequencias_treinamento = tokenizer.texts_to_sequences(dados_treinamento["text"])
sequencias_treinamento_padded = pad_sequences(sequencias_treinamento, maxlen=max_comprimento_texto, padding='post', truncating='post')

# Crie o modelo de classificação de texto
modelo = keras.Sequential([
    Embedding(input_dim=max_palavras, output_dim=16, input_length=max_comprimento_texto),
    Bidirectional(LSTM(64)),
    Dense(1, activation='sigmoid')
])

# Compile o modelo
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Treine o modelo com os dados de treinamento
modelo.fit(sequencias_treinamento_padded, dados_treinamento["label"], epochs=25, validation_split=0.2)

# Avalie o modelo com os dados de teste
sequencias_teste = tokenizer.texts_to_sequences(dados_teste["text"])
sequencias_teste_padded = pad_sequences(sequencias_teste, maxlen=max_comprimento_texto, padding='post', truncating='post')
resultado = modelo.evaluate(sequencias_teste_padded, dados_teste["label"])
print("Loss:", resultado[0])
print("Acurácia:", resultado[1])

# Use o modelo para prever se um texto é do autor específico ou não
texto_novo = ["Texto de teste a ser classificado"]
sequencia_nova = tokenizer.texts_to_sequences(texto_novo)
sequencia_nova_padded = pad_sequences(sequencia_nova, maxlen=max_comprimento_texto, padding='post', truncating='post')
previsao = modelo.predict(sequencia_nova_padded)
print("Probabilidade de ser do autor:", previsao[0][0])
```
